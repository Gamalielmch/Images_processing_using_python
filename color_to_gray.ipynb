```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.2989, 0.5870, 0.1140])
```

```python
img_c = plt.imread('im1.jpg')
img_g= rgb2gray(img_c)


fig1 = plt.figure(figsize = (15,8))

a = fig1.add_subplot(1, 2, 1)
plt.imshow(img_c, cmap='gray',vmin=0,vmax=2**8)
plt.axis('off');
a.set_title('Image color'+'size:'+ str(img_c.shape))

b = fig1.add_subplot(1,2, 2)
plt.imshow(img_g, cmap='gray',vmin=0,vmax=2**8)
plt.axis('off')
b.set_title('Image gray'+'size:'+ str(img_g.shape))
```

```python
import cv2
img1 = cv2.imread('im1.jpg')
img2 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)

fig1 = plt.figure(figsize = (15,8))

a = fig1.add_subplot(1, 2, 1)
plt.imshow(img1, cmap='gray',vmin=0,vmax=2**8)
plt.axis('off');
a.set_title('Image color'+'size:'+ str(img1.shape))

b = fig1.add_subplot(1,2, 2)
plt.imshow(img2, cmap='gray',vmin=0,vmax=2**8)
plt.axis('off')
b.set_title('Image gray'+'size:'+ str(img2.shape))
```
