```python
### Read MRI images 
import numpy as np
import matplotlib.pyplot as plt
from pydicom import dcmread
from pydicom.data import get_testdata_file
```

```python
mri=np.empty([256, 256,60],dtype=int)
ini=417
f= plt.figure(figsize = (15,40))

for i in range(60):
    nameF="MRI/I0000"+str(i+ini)+"_anon.dcm"
    ds = dcmread(nameF)
    mri[:,:,i]=ds.pixel_array
    f.add_subplot(12, 5, i + 1)
    plt.imshow(mri[:,:,i], cmap=plt.cm.gray)
    plt.axis('off');
plt.show(block=True)    
```
