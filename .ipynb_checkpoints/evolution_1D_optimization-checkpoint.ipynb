```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f_a =lambda t:((204165.5)/(330-2*t))+(10400/(t-20))
dx = lambda x, delta: (f_a(x+delta)-f_a(x))/delta
ddx = lambda x, delta: (f_a(x + delta) - 2 * f_a(x) + f_a(x - delta)) / (delta ** 2)
```

```python
##SECANTE
t = np.linspace(40,90,100)
delta=0.001
epsilon=1e-3
fu=dx(t,delta)
iterations=5
fig, ax = plt.subplots(nrows=iterations, ncols=1,figsize=(10, 70)) 
x1 = 90
x2 = 40
for i in range(iterations):
    ax[i].plot(t,fu)
    ax[i].grid(True, which='both')
    ax[i].axhline(y=0, color='k')
    ax[i].plot(x1,dx(x1,delta), color="red", marker="o",  markersize=7)
    ax[i].plot(x2,dx(x2,delta), color="red", marker="o",  markersize=7)
    x_values = [x1,x2]
    y_values = [dx(x1,delta),dx(x2,delta)]
    ax[i].plot(x_values, y_values)
    alpha = x2 - ((dx(x2,delta)*(x2-x1)))/(dx(x2,delta)-dx(x1,delta))
    ax[i].plot(alpha,dx(alpha,delta), color="green", marker="o",  markersize=7)
    ax[i].plot(alpha,0, color="yellow", marker="o",  markersize=7)
    ax[i].set_title("x: %f fx: %f f'x: %f" % (alpha,f_a(alpha),dx(alpha,delta)))
    if dx(alpha,delta)*dx(x1,delta)<0:
        x2=alpha
    else:
        x1 = alpha
    if abs(dx(alpha,delta))<epsilon:
        break
```

```python
##Newton
t = np.linspace(40,90,100)
delta=0.001
epsilon=1e-3
fu=dx(t,delta)
iterations=5
fig, ax = plt.subplots(nrows=iterations, ncols=1,figsize=(10, 70)) 
x1 = 90
x2 = 40
x=90
for i in range(iterations):
    ax[i].plot(t,fu, color="blue")
    ax[i].grid(True, which='both')
    ax[i].axhline(y=0, color='k')
    ax[i].plot(x,dx(x,delta), color="red", marker="o",  markersize=7)
    recta=dx(x,delta)+ddx(x,delta)*(t-x)
    # x_values = [x1,x2]
    # y_values = [dx(x1,delta),dx(x2,delta)]
    ax[i].plot(t, recta, color="cyan")
    x = x - dx(x,delta)/ddx(x,delta)
    ax[i].plot(x,dx(x,delta), color="green", marker="o",  markersize=7)
    ax[i].plot(x,0, color="yellow", marker="o",  markersize=7)
```
