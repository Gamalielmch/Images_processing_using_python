```python
import numpy as np
import matplotlib.pyplot as plt
import random
from celluloid import Camera 
import cv2
from IPython.display import HTML
```

```python
#Creating a Function.
def Npdf(x , mean , sd):
    prob_density = (1/(np.sqrt (2*np.pi)*sd)) * np.exp(-0.5*((x-mean)/sd)**2)
    return prob_density
```

```python
I = cv2.imread('im25.jpg')
I = cv2.cvtColor(I, cv2.COLOR_BGR2GRAY)
I=I.astype(float)
nf,nc=I.shape
N=nf*nc
I=I.ravel()
K=5
m=I[random.choices(np.arange(N),k=K)]
std=(np.ones(K,)*3)
phi=(np.ones(K,)/K)
gamma=np.zeros((N,K))
x=np.linspace(0,255,256)

for ite in range(20):
    fig=plt.figure
    plt.hist(I.ravel(),256,[0,256],density=True);
    gt=0
    for k in range(K):
        gt=gt+phi[k]*Npdf(x , m[k] , std[k])
        plt.plot(x,phi[k]*Npdf(x , m[k] , std[k]))
    plt.plot(x,gt,'--')

    for k in range(K):
        gamma[:,k]=Npdf(I , m[k] , std[k])
    suma=np.sum(gamma,axis=1)+1e-300
    gamma=gamma/suma.reshape(len(suma),1)
    suma2=np.sum(gamma,axis=0)
    phi=(suma2/N)
    for k in range(K):
        m[k]=np.sum(gamma[:,k]*I)/suma2[k]
        std[k]= np.sqrt(np.sum(gamma[:,k]*(I-m[k])**2)/suma2[k])
    
    plt.show()   
# animation = camera.animate() # animation ready    
# animation.save('test.mp4')
# HTML(animation.to_html5_video()) # displaying the animation
```
