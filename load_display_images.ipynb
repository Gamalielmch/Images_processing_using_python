```python
import cv2
import numpy as np
import tkinter as tk
import tkinter.filedialog as fd
import matplotlib.pyplot as plt

root = tk.Tk()
filez = fd.askopenfilenames(parent=root, title='Choose a files of object')
root.destroy()

It=cv2.imread(filez[0])
It = cv2.cvtColor(It, cv2.COLOR_BGR2GRAY)
[nf,nc]=It.shape
Phi=np.zeros((nf,nc,4))
Phi[:,:,0]=It
wx=np.zeros((nf,nc,4))
for i in range(1,4):
    It=cv2.imread(filez[0])
    It = cv2.cvtColor(It, cv2.COLOR_BGR2GRAY)
    It=It.astype(np.float64)
    Phi[:,:,i]=It

root = tk.Tk()
filez = fd.askopenfilenames(parent=root, title='Choose a file of plane')
root.destroy()
for i in range(0,4):
    It=cv2.imread(filez[0])
    It = cv2.cvtColor(It, cv2.COLOR_BGR2GRAY)
    It=It.astype(np.float64)
    wx[:,:,i]=It

Phi_w=np.arctan2(Phi[:,:,0]-Phi[:,:,1],Phi[:,:,2]-Phi[:,:,3])
print(Phi_w)
cv2.namedWindow("win", cv2.WINDOW_NORMAL)  
cv2.imshow("win", np.uint8(Phi_w))
cv2.waitKey(0) 
cv2.destroyAllWindows()
plt.imshow(Phi_w)
    
```
