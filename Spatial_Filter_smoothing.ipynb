```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
I = cv2.imread('im10.jpg')
I = cv2.cvtColor(I, cv2.COLOR_BGR2GRAY)
fig1=plt.figure(figsize = (4,4))
plt.imshow(I, cmap='gray',vmin=0,vmax=255)
plt.axis('off');
```

```python
If = cv2.blur(I,(3, 3))
```

```python
fig1=plt.figure(figsize = (4,4))
plt.imshow(If, cmap='gray',vmin=0,vmax=255)
plt.axis('off');
```
