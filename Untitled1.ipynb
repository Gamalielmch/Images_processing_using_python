```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import random
```

```python
I = cv2.imread('im25.jpg')
I = cv2.cvtColor(I, cv2.COLOR_BGR2RGB)
nf,nc,nch=I.shape
print(nf,nc,nch)
plt.imshow(I,  cmap='gray')
```

```python
Ir=I[:,:,0]
Ig=I[:,:,1]
Ib=I[:,:,2]
fig1=plt.figure(figsize = (18,7))
plt.imshow(np.c_[Ir,Ig,Ib],  cmap='jet')
```

```python
Ir=Ir.ravel().astype(float)
Ig=Ig.ravel().astype(float)
Ib=Ib.ravel().astype(float)
K=3
mr = [Ir[random.randint(0, len(Ir))] for i in range(K)]
mg = [Ig[random.randint(0, len(Ig))] for i in range(K)]
mb = [Ib[random.randint(0, len(Ib))] for i in range(K)]
print(mr,mb,mg)
```

```python
distancia=np.zeros((nc*nf,K))
for ite in range(30):
    for k in range(K): 
        distancia[:,k]=np.sqrt((Ir-mr[k])**2+(Ig-mg[k])**2+(Ib-mb[k])**2)
    assig=np.argmin(distancia,axis=1)
    for k in range(K):
            mr[k]=np.mean(Ir[assig==k])
            mg[k]=np.mean(Ig[assig==k])
            mb[k]=np.mean(Ib[assig==k])
    print("medias rojo: %2.2f %2.2f %2.2f" %(mr[0],mr[1],mr[2]))
    print("medias verde: %2.2f %2.2f %2.2f" %(mg[0],mg[1],mg[2]))
    print("medias azul: %2.2f %2.2f %2.2f" %(mb[0],mb[1],mb[2]))
    print("_____________________")
```

```python
I_seg_r=Ir*1
I_seg_g=Ig*1
I_seg_b=Ib*1
for k in range(K): 
    distancia[:,k]=np.sqrt((Ir-mr[k])**2+(Ig-mg[k])**2+(Ib-mb[k])**2)
assig=np.argmin(distancia,axis=1)
for k in range(K):
    I_seg_r[assig==k]=mr[k]
    I_seg_g[assig==k]=mg[k]
    I_seg_b[assig==k]=mb[k]
I_seg=I*1
I_seg[:,:,0]=I_seg_r.reshape(nf,nc)    
I_seg[:,:,1]=I_seg_g.reshape(nf,nc) 
I_seg[:,:,2]=I_seg_b.reshape(nf,nc) 
plt.imshow(I_seg, cmap='gray')
```

```python
I = cv2.imread('im24.jpg')
I = cv2.cvtColor(I, cv2.COLOR_BGR2GRAY)
plt.hist(I.ravel(),256,[0,256]);
var_max=0
umbral1=2
umbral2=3
for i in range(2,252):
    for j in range(i+1,253):
        var1=np.var(I[I<i])
        It=I[I>=i]
        It=It[It<j]
        var2=np.var(It)
        var3=np.var(I[I>=j])
        vart=var1+var2+var3
        if vart>var_max:
            var_max=vart
            umbral1=i
            umbral2=j
        
print(umbral1,umbral2)       
```
