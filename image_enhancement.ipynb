```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline
```

```python
I = cv2.imread('im7.jpg')
I = cv2.cvtColor(I, cv2.COLOR_BGR2GRAY)
fig1=plt.figure(figsize = (4,4))
plt.imshow(I, cmap='gray',vmin=0,vmax=255)
plt.axis('off');
```

```python
inmax = np.amax(I)
inmin = np.amin(I)
print('min=',inmin,'max=',inmax)
Size=I.shape
```

## Linear enhancement

```python
Im=255*((I-inmin)/(inmax-inmin))
inmax = np.amax(Im)
inmin = np.amin(Im)
print('min=',inmin,'max=',inmax)
```

```python
fig2=plt.figure(figsize = (8,8))
a = fig2.add_subplot(1, 2, 1)
plt.imshow(I, cmap='gray',vmin=0,vmax=255)
plt.axis('off');

b = fig2.add_subplot(1, 2, 2)
plt.imshow(Im, cmap='gray',vmin=0,vmax=255)
plt.axis('off');
```

```python
## log and exp enhancement
```

```python
x=np.linspace(0, 255, num=256)
c=255/np.log(1+255)
s=c*np.log(1+x)
plt.plot(x,s, label="f\'(x)" , color = "red" )
```

```python
c=255/np.log(1+255)
Il=c*np.log(1+I)
Il=Il.astype(np.uint8)
inmax = np.amax(Il)
inmin = np.amin(Il)
print('min=',inmin,'max=',inmax)
```
