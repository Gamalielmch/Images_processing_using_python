```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
NPDF= lambda x,m,sd: (1/(np.sqrt (2*np.pi)*sd)) * np.exp(-0.5*((x-m)/sd)**2)
```

```python
I = cv2.imread('im25.jpg',0)
I = I.ravel().astype(float)
k = 5
m = np.random.choice(np.unique(I), k,  replace = False)
std=np.ones((k,))*5
phi=np.ones((k,))/k
N=I.shape[0]
gamma=np.zeros((N,k))
fig=plt.figure
plt.hist(I,256,[0,256],density=True);
x=np.linspace(0,255,256)
for i in range(k):
    plt.plot(x,phi[i]*NPDF(x,m[i],std[i]))
plt.show()  

for j in range (k):
    gamma[:,j]=phi[j]*NPDF(I,m[j],std[j])
suma=np.sum(gamma,axis=1)
suma=np.repeat(suma, k).reshape(N,k)
gamma=gamma/suma
print(std)
for j in range(k):
    phi[j]=np.sum(gamma[:,j])
    m[j]=np.dot(gamma[:,j],I)/phi[j]
    std[j]=np.sqrt(np.sum(gamma[:,j]*(I-m[j])**2)/phi[j])
    phi[j]=phi[j]/N


fig2=plt.figure    
plt.hist(I,256,[0,256],density=True);
x=np.linspace(0,255,256)
for i in range(k):
    plt.plot(x,phi[i]*NPDF(x,m[i],std[i]))
plt.show()  


```
