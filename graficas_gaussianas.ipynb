```python
import numpy as np 
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
from mpl_toolkits.mplot3d import Axes3D
```

```python
#Creating a Function.
def normal1D(x , mean , sd):
    prob_density = (1/(np.sqrt (2*np.pi)*sd)) * np.exp(-0.5*((x-mean)/sd)**2)
    return prob_density
```

```python
x=np.linspace(0,255,255)
mu=80
std=5
fx=normal1D(x , mu , std)
plt.plot(x,fx)
```

```python
%matplotlib widget
x1=np.linspace(0,255,500)
x2=np.linspace(0,255,500)
mu=np.array([[128,128]]).reshape(2,)
cova=np.array([[500,300],[300,500]])
fx=np.zeros((len(x1),len(x2)))
for i in range(len(x1)):
    for j in range(len(x2)):
        fx[i,j]=multivariate_normal.pdf([x1[i],x2[j]], 
                                        mean=mu , cov=cova)
X2, X1 = np.meshgrid(x1, x2)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X2,X1, fx,
                       linewidth=0, antialiased=False, cmap='jet')
ax.set_xlabel('x1',fontsize=12)
ax.set_ylabel('x2',fontsize=12)
ax.set_zlabel('fx',fontsize=12)
```
